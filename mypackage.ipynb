```python
import numpy as np
import pylab as plt
from ipywidgets import interact, fixed
from PIL import Image

originalSize = (100,100)
newSize = (64,64)
imarray = np.random.rand(100,100,3) * 255
image = Image.fromarray(imarray.astype('uint8')).convert('RGBA')
image.save('result_image.png') 

def imshowResized(X, imgresize=None):
    """
    You should create a way to resize an image from an array X. 
    The use of widgets is optional but you can take a look to interact. 
    We should be able to install this package in Google Colab from your Git repo.
    """
   
    resized = np.resize(X, imgresize)
    return resized



figure, (number1, number2) = plt.subplots(1,2)

number1.imshow(image)
number1.title.set_text("original image: ")
number2.imshow(imshowResized(image, newSize))
number2.title.set_text("resized image: ")



 
```
